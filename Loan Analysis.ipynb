```python
# 5.Plot a bar chart using loan status to visualize the approval rate for each occupation status and that show the approval rate for different 
# credit history years
# Calculate approval rate for occupation
total_counts = loan.groupby('occupation_status')['loan_status'].count()
approved_counts = loan[loan['loan_status'] == 1].groupby('occupation_status')['loan_status'].count()
approved_counts = approved_counts.reindex(total_counts.index, fill_value=0)
approval_rate = (approved_counts / total_counts).reset_index(name='approval_rate')
loan = loan.merge(approval_rate, on='occupation_status', how='left')
loan.head(1)
```

#### This tells us how many loans were approved in each credit_history_years group. It highlights which groups are more likely to get approved.

```python
# Calculate approval rate for credit history
total_ch = loan.groupby('credit_history_years')['loan_status'].count()
approved_ch = loan[loan['loan_status'] == 1].groupby('credit_history_years')['loan_status'].count()
approved_ch = approved_ch.reindex(total_ch.index, fill_value=0)
credit_history_rate = (approved_ch / total_ch).reset_index(name='credit_history_approval_rate')
loan = loan.merge(credit_history_rate, on='credit_history_years', how='left')
loan.head(1)
```

#### Credit history affects loan approval: People with longer credit histories are more likely to get their loans approved.

```python
plt.figure(figsize=(12,6))
h=sns.barplot(data=loan,x='occupation_status',y='approval_rate',hue='credit_history_approval_rate')
for i in h.containers:
    h.bar_label(i)
plt.title('Credit History - Occupation Status vs Approval Rate')
plt.xlabel('Occupation Status')
plt.ylabel('Approval Rate')
plt.legend()
```

#### The plot shows which occupation groups with specific credit history lengths are more likely to get approved.

```python
# 6.Build a heatmap showing correlations between: annual income loan amount interest rate credit score debt-to-income ratio.
#Correlation Heatmap (Matplotlib Only)

col = ['annual_income','loan_amount','interest_rate','credit_score','debt_to_income_ratio']
# columns to calculate the correlation matrix
corr_matrix = loan[col].corr()
# Plot Graph
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',cbar_kws={'label': 'Correlation Coefficient'})
plt.title("Correlation Heatmap of Financial Variables")
plt.tight_layout()
```

#### Conclusion : 
* Loan Amount and Annual Income have the strongest positive correlation (0.75), which is expected as people with higher income can borrow larger amounts.
* Loan Amount and Credit Score also have a positive correlation (0.57).
* Interest Rate shows a negative correlation with both Annual Income (-0.20) 
and Credit Score (-0.19), meaning higher income/better credit tends to result in lower interest rates.

```python
# 7.How does loan approval vary across income categories?
# Create Income Category (Low / Medium / High)
category = ['Low', 'Medium', 'High']
category_type = pd.CategoricalDtype(categories=category,ordered=True)
loan['income_category'] = pd.qcut(loan['annual_income'],q=3,labels=category,duplicates='drop').astype(category_type)
approval_rate = loan.groupby('income_category',observed=False)['loan_status'].mean().reset_index()
approval_rate.columns = ['income_category', 'approval_rate_raw']
approval_rate['approval_rate'] = approval_rate['approval_rate_raw'] * 100
```

```python
plt.figure(figsize=(12,6))
a=sns.barplot(data=approval_rate,x='income_category',y='approval_rate',order=category,hue='income_category')
for i in a.containers:
    a.bar_label(i)
plt.title('Loan Approval Across Income Categories')
plt.xlabel('Income Category')
plt.ylabel('Approval Rate')
```

#### Conclusion: 
* The annual income of an applicant is one of the most important factors determining whether a loan is approved.
* Loan approval increases significantly as the borrower's income category moves from Low to High.

```python
# 8.Which credit score group borrows the highest loan amount?
# Create Credit Score Category (Poor, Fair, Good, Excellent)

loan["credit_category"] = pd.cut(loan["credit_score"],bins=[300, 580, 670, 740, 850],labels=["Poor", "Fair", "Good", "Excellent"])

avg_loan = loan.groupby("credit_category",observed=False)["loan_amount"].mean().reset_index()

plt.figure(figsize=(12,6))
z=sns.barplot(data=avg_loan,x="credit_category",y="loan_amount",hue='credit_category',palette='magma')
for i in z.containers:
    z.bar_label(i)
plt.title("Average Loan Amount by Credit Score Category")
plt.xticks(rotation=45)
plt.xlabel("Credit Score Category")
plt.ylabel("Average Loan Amount")
plt.show()
```

#### Conclusion :
Applicants/Customers who have Good and Excellent credit scores tend to borrow higher loan amounts compared to Poor and Fair credit groups.

```python
# 9.How does EMI differ for approved vs. rejected applicants?
# Create Monthly EMI Column
# EMI = loan_amount * interest_rate / 100

# calclution of EMI
loan["EMI"] = loan["loan_amount"] * (loan["interest_rate"] / 100)

# Plotting
plt.figure(figsize=(12,5))
e=sns.boxplot(data=loan,x="loan_status",y="EMI",hue='loan_status',palette='viridis')
plt.title("EMI by Loan Status (Approved vs Rejected)")
plt.xlabel("Loan Status (0 = Rejected, 1 = Approved)")
plt.ylabel("EMI Amount")
```

#### Conclusion :
* Rejected applicants generally show higher EMI amounts, meaning their EMI burden is large compared to their financial profile.

* Approved applicants have lower or manageable EMI, which makes them safer for lenders.

```python
# 10.What is the proportion of approved vs rejected loans?
# Column used: loan_status
# 1= Approved #0= Rejected

counts = loan["loan_status"].value_counts()
print(counts)
# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(counts,labels=["Approved (1)","Rejected (0)"],autopct="%1.1f%%",startangle=90,explode=(0.05,0))

plt.title("Proportion of Approved vs Rejected Loans")
```

#### Conclusion :
* The pie chart shows what percentage of loan applications were approved vs rejected.
* If the “Approved” slice is larger → your dataset has more successful applications.
* If “Rejected” is larger → loan approval criteria is showing very strict in given dataset.

```python
# 11.What is the percentage distribution of different loan intents?
# Column used: loan_intent (education, business, medical, personal, etc.)

li = loan["loan_intent"].value_counts(normalize=True) * 100
print(li)
# Pie Chart
plt.figure(figsize=(12,7))
plt.pie(li,labels=li.index,autopct="%1.1f%%")
plt.title("Percentage Distribution of Loan Intents",fontsize=14)
```

#### Conclusion :
